```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
cores = [x for x in range(2, 22+2, 2)]

times1 = [2178, 1709, 1618, 1495, 1510, 1437, 1666, 2363, 4666, 1274, 713]
 
times2 = [2078, 1641, 1565, 1406, 1389, 1355, 1609, 2348, 1141, 1244, 1297]



plt.figure(num=None, figsize=(8, 6), dpi=160, facecolor='w', edgecolor='k')
plt.plot(cores, times1, 'x-', label='run 1')
plt.plot(cores, times2, 'h-', label='run 2')
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.xticks(np.arange(2, 22+2, 2.0))
plt.title('Effect of increasing parallelisation on clock time', fontsize=18)
plt.xlabel('n cores', fontsize=16)
plt.ylabel('clocktime (s)', fontsize=16)
plt.legend(loc='upper left', fontsize=14)
plt.show()
```
